## <font color='green'>**2.5 Identificação de Problemas nos Dados**</font>

### <font color='cyan'> Q16: Inspecione o conjuntos de dados para verificar se há:</font>
- #### <font color='cyan'> Dados Ausentes;</font>
- #### <font color='cyan'> Valores Atípicos (via IQR);</font>
- #### <font color='cyan'> Valores Duplicados;</font>
### <font color='cyan'> Reporte sua(s) descoberta(s) a seguir.</font>

```python
# 1. Checking for Missing Data
print("\n--- Checking for Missing Data ---")
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
if not missing_values.empty:
    print("Columns with missing values:\n", missing_values)
else:
    print("No missing values found in the dataset.")

# 2. Checking for Outliers (using IQR for numerical columns)
print("\n--- Checking for Outliers (IQR Method) ---")
numerical_cols = df.select_dtypes(include=['number']).columns

outlier_found = False
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    if not outliers.empty:
        outlier_found = True
        print(f"Column '{col}' has {len(outliers)} outliers. (Min: {df[col].min()}, Max: {df[col].max()}, Q1: {Q1}, Q3: {Q3}, Lower Bound: {lower_bound:.2f}, Upper Bound: {upper_bound:.2f})")
if not outlier_found:
    print("No significant outliers detected in numerical columns based on IQR.")

# 3. Checking for Duplicate Values
print("\n--- Checking for Duplicate Values ---")
duplicates_count = df.duplicated().sum()
if duplicates_count > 0:
    print(f"The dataset has {duplicates_count} duplicate rows.")
else:
    print("No duplicate rows found in the dataset.")
```

### <font color='cyan'> Q17: Caso um problema tenha sido encontrado, proponha uma solução para ele.</font>
### <font color='cyan'> Exiba o conjunto de dados antes e depois da solução, enfatizando as mudanças.</font>

```python
df_cleaned = df.copy()

# --- 1. Handling Duplicate Values ---
print("--- Handling Duplicate Values ---")
print(f"Before removing duplicates: {df_cleaned.duplicated().sum()} duplicate rows.")
df_cleaned.drop_duplicates(inplace=True)
print(f"After removing duplicates: {df_cleaned.duplicated().sum()} duplicate rows.")
print(f"New dataset shape after removing duplicates: {df_cleaned.shape}")

# --- 2. Handling Missing Values ---
print("\n--- Handling Missing Values ---")
missing_cols = ['workclass', 'occupation', 'native-country']
print("Missing values before imputation:")
print(df_cleaned[missing_cols].isnull().sum())

for col in missing_cols:
    mode_value = df_cleaned[col].mode()[0]
    df_cleaned[col].fillna(mode_value, inplace=True)

print("\nMissing values after mode imputation:")
print(df_cleaned[missing_cols].isnull().sum())

# --- 3. Handling Outliers (Capping using IQR) ---
print("\n--- Handling Outliers (Capping using IQR) ---")
numerical_cols_with_outliers = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

print("Min/Max before outlier capping:")
print(df_cleaned[numerical_cols_with_outliers].min())
print(df_cleaned[numerical_cols_with_outliers].max())

for col in numerical_cols_with_outliers:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap values
    df_cleaned[col] = df_cleaned[col].clip(lower=lower_bound, upper=upper_bound)

print("\nMin/Max after outlier capping:")
print(df_cleaned[numerical_cols_with_outliers].min())
print(df_cleaned[numerical_cols_with_outliers].max())

print("\n--- Cleaned Dataset Sample (First 5 rows) ---")
print(df_cleaned.head())
```
